```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
city_df=pd.read_csv(city_data_to_load)
rides_df=pd.read_csv(ride_data_to_load)
dataset_table=pd.merge(city_df,rides_df, on="city")
dataset_table.head()
```

```python
dataset_table.type.unique() #identifying unique types
```

```python
# calcualting variables
Rides=dataset_table.groupby("city")["ride_id"]
Total_Number_of_Rides=Rides.count()  # Total number of rides
fare=dataset_table.groupby("city")["fare"]
fare_total=fare.sum()
Average_Fare=round(fare_total/Total_Number_of_Rides,2) # average fare rounding to two decimals 
drivers=dataset_table.groupby("city")["driver_count"]
Drivers_count=drivers.mean() # count od drivers per city
types=city_df.set_index("city")["type"]
```

```python
#creating a new dataframe with the required information to plot 
data_set2=pd.DataFrame({"Total Number of Drivers":Drivers_count, "Average Fare":Average_Fare,"Total Number of Rides":Total_Number_of_Rides, "Type of City":types})
#data_set2.head()
```

```python
#setting the dataset 
Urban=data_set2.loc[data_set2["Type of City"]=="Urban"]
Suburban=data_set2.loc[data_set2["Type of City"]=="Suburban"]
Rural=data_set2.loc[data_set2["Type of City"]=="Rural"]
```

```python
#creating a nice very nice plot
Fig1=plt.scatter(Urban["Total Number of Rides"],Urban["Average Fare"], color="lightcoral",edgecolor="black", s=Urban["Total Number of Drivers"]*10,label="Urban", alpha=0.7, linewidth=2) 
Fig2=plt.scatter(Suburban["Total Number of Rides"],Suburban["Average Fare"], color="lightskyblue",edgecolor="black", s=Suburban["Total Number of Drivers"]*10,label="Urban", alpha=0.7, linewidth=2) 
Fig3=plt.scatter(Rural["Total Number of Rides"],Rural["Average Fare"], color="Gold",edgecolor="black", s=Rural["Total Number of Drivers"]*10,label="Urban", alpha=0.7, linewidth=2) 
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Ride (Per City)")
plt.ylabel("Average Fare($)")
plt.text(45,35,"Note:Circle size correlates with driver count per city.")
plt.legend((Fig1,Fig2,Fig3),("Urban","Suburban","Rural" ))
plt.grid(True)

plt.show()
```

```python
#% of Total Fares by City Type

City_Fare= dataset_table.groupby("type")["fare"].sum()


#% of Total Rides by City Type
Rides= dataset_table.groupby("type")["ride_id"].count()

#% of Total Drivers by City Type
Drivers= dataset_table.groupby("type")["driver_count"].mean()
```

```python
# pie % of Total Fares by City Type
explode=(0,0,0.1)
colors=["gold","lightskyblue","lightcoral"]
label="Rural","Suburban","Urban"
plt.pie(City_Fare, labels=label, autopct='%1.1f%%',colors=colors, explode= explode,shadow=True, startangle=90 )
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.legend()
plt.show()
```

```python
# pie % of Total Rides by City Type
explode=(0,0,0.1)
colors=["gold","lightskyblue","lightcoral"]
label="Rural","Suburban","Urban"
plt.pie(Rides, labels=label, autopct='%1.1f%%',colors=colors, explode= explode,shadow=True, startangle=90 )
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.legend()
plt.show()
```

```python
# pie % of Total Drivers by City Type
explode=(0,0,0.1)
colors=["gold","lightskyblue","lightcoral"]
label="Rural","Suburban","Urban"
plt.pie(Drivers, labels=label, autopct='%1.1f%%',colors=colors, explode= explode,shadow=True, startangle=90 )
plt.axis("equal")
plt.title(" % of Total Drivers by City Type")
plt.legend()
plt.show()
```
